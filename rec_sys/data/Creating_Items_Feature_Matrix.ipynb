```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
pd.set_option('display.max_columns', 100)
```

```python
collection_belongings = pd.read_csv('collection_belongings.csv')
course_belongings = pd.read_csv('course_belongings.csv')
lang_feature_matrix = pd.read_csv('language_feature_matrix.csv')
lessons_info = pd.read_csv('lessons.csv')
userlessons_info = pd.read_csv('user_lessons.csv')
```

```python
lang_feature_matrix
```

```python
cos_sim = pd.DataFrame(cosine_similarity(lang_feature_matrix.values[:,1:]),columns=lang_feature_matrix.Language.values)
cos_sim.insert(loc=0,column='',value=lang_feature_matrix.Language.values)
cos_sim
```

```python
all_lessons = list(userlessons_info.Id) + list(lessons_info.Id)
len(all_lessons)
```

```python
lessons_info.shape
```

```python
userlessons_info.shape
```

```python
UL_temp = pd.merge(userlessons_info,collection_belongings[['Id','CollectionName']],on='Id',how='left')
UL_temp.head(10)
```

```python
L_temp = pd.merge(lessons_info,course_belongings[['Id','CourseName']],on='Id',how='left')
L_temp.head(10)
```

```python
UL_temp = UL_temp.rename(columns={'CollectionName':'BelongsTo'})
L_temp = L_temp.rename(columns={'CourseName':'BelongsTo'})
```

```python
print(UL_temp.columns)
print(L_temp.columns)
```

```python
data = pd.concat([UL_temp,L_temp])
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
data.shape
```

```python
id_to_language = data[['Id','Language']].copy()
id_to_language = id_to_language.drop_duplicates(subset='Id')
print(id_to_language.shape)
```

```python
# Getting dummies
belongings = pd.get_dummies(data.BelongsTo)
data = data.drop(['BelongsTo','Language'],axis=1)
data = data.join(belongings)
data.head(20)
```

```python
data[data.Id=='U704']
```

```python
new_data = data.groupby(['Id']).sum().reset_index()
new_data.head(20)
```

```python
new_data[new_data.Id=='U704']
```

```python
print(new_data.shape)
```

```python
new_data = pd.merge(new_data,id_to_language,on='Id',how='inner')
new_data.head(20)
```

```python
new_data = pd.merge(new_data,lang_feature_matrix,on='Language',how='left')
new_data = new_data.drop(['Language'],axis=1)
new_data.head(20)
```

```python
new_data[new_data.Id=='U691']
```

```python
new_data.fillna(0,inplace=True)
```

```python
new_data[new_data.Id=='U691']
```

```python
view_count_min = np.min(new_data.ViewCount)
view_count_max = np.max(new_data.ViewCount)
comment_reply_min = np.min(new_data.CommentReply)
comment_reply_max = np.max(new_data.CommentReply)
```

```python
new_data.describe()
```

```python
new_data.ViewCount = new_data.ViewCount.apply(lambda x: (x-view_count_min)/(view_count_max-view_count_min))
new_data.CommentReply = new_data.CommentReply.apply(lambda x: (x-comment_reply_min)/(comment_reply_max-comment_reply_min))
```
